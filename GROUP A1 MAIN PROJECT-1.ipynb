## IMPORTING THE LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##  IMPORTING THE DATA

```python
df = pd.read_csv('United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv')
```

## DATAFRAME FORMATTING 

```python
df.submission_date = pd.to_datetime(df.submission_date)   #CONVERTING THE Dtype OF "submission_date" TO DATETIME
```

```python
df.sort_values(by=['submission_date'], inplace=True)      #SORTING THE DATA FRAME AS ACENDING ORDER OF DATE
df
```

```python
df.drop(['created_at','consent_cases','consent_deaths'], axis=1, inplace=True)    #DROPPING THE LESS USEFULL DATA
```

```python
df.reset_index(inplace=True)       #RESETTING THE INDEXING
```

```python
df.drop(['index'], axis=1, inplace=True)     #DELETING THE OLD INDEXING 
```

```python
df.fillna(value=0, inplace = True)     #REPLACING NAN WITH 0
```

## GROUPING DATA 

```python
df_date =df.groupby('submission_date').sum()  #GROUPPING THE DATAFRAME ACCORDING THE DATE OF SUBMISSION
```

```python
df_state =df.groupby('state').sum()           #GROUPPING THE DATAFRAME ACCORDING TO THE STATES
```

## ADDING NEWS COLUMNS

```python
df_date['pos_rate'] = ((df_date['conf_cases']) / (df_date['tot_cases']))  * 100    #POSITIVITY RATE
```

```python
df_date['fatality_rate'] = ((df_date['tot_death'] / df_date['conf_cases'])) * 100  #FATALITY RATE
```

```python
df_state['pos_rate'] = ((df_state['conf_cases']) / (df_state['tot_cases']))  * 100    #POSITIVITY RATE
```

```python
df_state['fatality_rate'] = ((df_state['tot_death'] / df_state['conf_cases'])) * 100  #FATALITY RATE
```

## PLOTTING THE DATA

```python
plt.figure(figsize=(12, 6), dpi=450)
plt.plot(df_date.index, df_date["new_case"],linewidth=0.8, color='#f41635')
plt.xlabel('Date')
plt.ylabel('Number')
plt.title('New Covid Cases')
```

```python
plt.figure(figsize=(12, 6), dpi=450)
plt.plot(df_date.index, df_date["new_death"],linewidth=0.8, color='#f41635')
plt.xlabel('Date')
plt.ylabel('Number')
plt.title('New Covid Deaths')
```

```python
plt.figure(figsize=(12, 6), dpi=450)
plt.plot(df_date.index, df_date["conf_cases"], linewidth=1.2, color='#ffa200')
plt.xlabel('Date')
plt.ylabel('Number')
plt.title('Total Covid Cases')
```

```python
plt.figure(figsize=(12, 6), dpi=450)
plt.plot(df_date.index, df_date["tot_death"], linewidth=1.2, color='#ffa200')
plt.xlabel('Date')
plt.ylabel('Number')
plt.title('Total Covid Deaths')
```

```python
plt.figure(figsize=(12, 6), dpi=450)
plt.plot(df_date.index, df_date["tot_cases"], color='y', linewidth=1.2, label='Total Tests')
plt.plot(df_date.index, df_date["conf_cases"], color='r', linewidth=1.2, label='Cases')
plt.xlabel('Date')
plt.ylabel('Number')
plt.title('Positivity Rate = Cases/Tests')
plt.legend()
```

```python
plt.figure(figsize=(12, 6), dpi=450)
plt.plot(df_date.index, df_date["pos_rate"], linewidth=1.2, color='r')
plt.title('Positivity Rate')
plt.xlabel('Date')
plt.ylabel('Percentage(%)')
```

```python
plt.figure(figsize=(12, 6), dpi=450)
plt.plot(df_date.index, df_date["tot_death"], color='r', linewidth=1.2, label='Deaths')
plt.plot(df_date.index, df_date["conf_cases"], color='y', linewidth=1.2, label='Total cases')
plt.title('Fatality Rate = Deaths/Cases')
plt.xlabel('Date')
plt.ylabel('Number ')
plt.legend()
```

```python
plt.figure(figsize=(12, 6), dpi=450)
plt.plot(df_date.index, df_date["fatality_rate"], linewidth=1.2,color='r')
plt.title('Fatality Rate')
plt.xlabel('Date')
plt.ylabel('Percentage(%)')
```

```python
plt.figure(figsize=(12, 6), dpi=450)
sns.barplot(data=df_state, x=df_state.index, y="tot_death")
plt.xticks(rotation=90)
plt.xlabel('US States')
plt.ylabel('Number')
plt.title('City Wise Deaths')
```
